```python
import pandas as pd
```

```python
full_data = pd.read_csv('../csv/half_processed_wins.csv')
```

```python
def filter(names, df):
    mask = df['Players 1'].str.contains('|'.join(names))
    mask |= df['Players 2'].str.contains('|'.join(names))
    return df[mask]
```

```python

def player_stats(player_name):
    # Filtering based on player name
    df = filter([player_name], full_data)
    df_player1 = df[df['Players 1'].str.contains(player_name)]
    df_player2 = df[df['Players 2'].str.contains(player_name)]
    
    wins_as_player1 = df_player1[df_player1['Score 1'] > df_player1['Score 2']]
    wins_as_player2 = df_player2[df_player2['Score 2'] > df_player2['Score 1']]
    total_wins = len(wins_as_player1) + len(wins_as_player2)
    
    losses_as_player1 = df_player1[df_player1['Score 1'] < df_player1['Score 2']]
    losses_as_player2 = df_player2[df_player2['Score 2'] < df_player2['Score 1']]
    
    # Average Scores on Wins and Losses
    avg_score_wins = (wins_as_player1['Score 1'].sum() -wins_as_player1['Score 2'].sum() + wins_as_player2['Score 2'].sum() -wins_as_player2['Score 1'].sum()) / total_wins if total_wins else 0
    total_losses = len(losses_as_player1) + len(losses_as_player2)
    avg_score_losses = (losses_as_player1['Score 1'].sum() -losses_as_player1['Score 2'].sum() + losses_as_player2['Score 2'].sum() -losses_as_player2['Score 1'].sum()) / total_losses if total_losses else 0
    
    # Points stats
    total_points_scored = df_player1['Score 1'].sum() + df_player2['Score 2'].sum()
    total_points_conceded = df_player1['Score 2'].sum() + df_player2['Score 1'].sum()
    
    print(f"Player: {player_name}")
    print(f"Total Wins: {total_wins}")
    print(f"Total Losses: {total_losses}")
    print(f"Average Score on Wins: {avg_score_wins}")
    print(f"Average Score on Losses: {avg_score_losses}")
    print(f"Total Points Scored: {total_points_scored}")
    print(f"Total Points Conceded: {total_points_conceded}")
    print(df['Division'].value_counts())
    print(df['Tournament Name'].value_counts())
```

```python
player_stats('fredric hinkle')
```
